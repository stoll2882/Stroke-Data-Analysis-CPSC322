# Data Distributions

To show the distributions of the attributes we will be using from the stroke data.



```python
import mypytable
from mypytable import MyPyTable
import importlib
import utils
import plot_utils
```

```python
stroke_table = MyPyTable()
stroke_table.load_from_file("chosen_data.csv");
```

```python
gender_col = utils.get_col_no_clean(stroke_table.data, stroke_table.column_names, "gender")
utils.convert_to_numeric(gender_col)
```

```python
gender_values, gender_counts = utils.get_frequencies(stroke_table.data, stroke_table.column_names, "gender")

gender_vals = ["Female", "Male"]
percents = []
num_female = sum(gender_col)

num_entries = len(gender_col)
num_male = num_entries - num_female

female_percent = num_female / num_entries
percents.append(female_percent)
male_percent = num_male / num_entries
percents.append(male_percent)

gender_title = "Frequency Distribution by Gender"
plot_utils.pie_chart(gender_vals, percents, gender_title)

#plot_utils.bar_chart(gender_values, gender_counts, gender_title)
```

# Distribution by Age Group

```python
age_values, age_counts = utils.get_frequencies(stroke_table.data, stroke_table.column_names, "age")
age_title = "Frequency Distribution by Age Group"
plot_utils.bar_chart(age_values, age_counts, age_title)
```

This graph shows us that our age groups are skewed to the right with a peak in the range of 70-80 years old.

# Marriage Status Distribution 

```python
marry_col = utils.get_col_no_clean(stroke_table.data, stroke_table.column_names, "ever_married")
utils.convert_to_numeric(marry_col)

marry_values, marry_counts = utils.get_frequencies(stroke_table.data, stroke_table.column_names, "ever_married")

married_vals = ["Yes", "No"]
percents = []
num_married = sum(marry_col)

num_entries = len(marry_col)
num_not_married = num_entries - num_married

married_percent = num_married / num_entries
percents.append(married_percent)
never_percent = num_not_married / num_entries
percents.append(never_percent)

marry_title = "Frequency Distribution by Marriage Status"
plot_utils.pie_chart(married_vals, percents, marry_title)
```
